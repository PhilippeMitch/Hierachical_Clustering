```
# import the librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering
```

```
# Load the dataset
data = pd.read_csv('credit_bank.csv')
```

```
data.head()
```

```
fusions = linkage(data, method='complete', metric='euclidean')
```

```
# Find the number of optimum cluster
dendrogram(fusions, leaf_rotation=90, leaf_font_size=6)
plt.title('Dendogram')
plt.xlabel('Credit Applicant')
plt.ylabel('Euclidean distances')
plt.show()
```

```
# Fit our Hierachiral Clustering with our data
h_clustering = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage='complete')
y_clustering = h_clustering.fit_predict(data)
```

```
data = data.values
```

```
plt.scatter(data[y_clustering == 0, 0], data[y_clustering == 0, 1], s = 100, c = 'cyan', label = 'cluster 1')
plt.scatter(data[y_clustering == 1, 0], data[y_clustering == 1, 1], s = 100, c = 'red', label = 'cluster 2')
plt.scatter(data[y_clustering == 2, 0], data[y_clustering == 2, 1], s = 100, c = 'green', label = 'cluster 3')
plt.title('Cluster for Credit Applicant')
plt.xlabel('Savings in Thousand')
plt.ylabel('Score_bank')
plt.legend()
plt.show()
```
